# **Python Packages:**

## **1. What is a Package in Python?**

* A **package** is a way of organizing related Python code into **modules** (files) and **sub-packages** (folders).
* Itâ€™s essentially a **directory with an `__init__.py` file** (in Python 3.3+, this file can be optional, but its presence makes the folder explicitly a package).

**Example structure:**

```
my_package/
    __init__.py
    module1.py
    module2.py
    sub_package/
        __init__.py
        sub_module.py
```
---

## **2. Difference Between Modules and Packages**

* **Module** â†’ A single `.py` file with Python code.
* **Package** â†’ A collection of modules in a directory (with `__init__.py`).

Example:



```python
# main.py
import module1
print(module1.greet("Alice"))
```

## **3. Creating a Package**

Letâ€™s create a package called `math_utils`.

```
math_utils/
    __init__.py
    arithmetic.py
    geometry.py
```

**arithmetic.py**

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
```

**geometry.py**

```python
import math

def area_circle(radius):
    return math.pi * radius * radius
```

****init**.py**

```python
# Expose selected functions at package level
from .arithmetic import add, subtract
from .geometry import area_circle
```

Now you can use:

```python
import math_utils

print(math_utils.add(5, 3))       # 8
print(math_utils.area_circle(2)) # 12.566...
```

---

## **4. Importing from Packages**

Different ways to import:

```python
import math_utils
print(math_utils.add(2, 3))

from math_utils import add
print(add(2, 3))

from math_utils.arithmetic import subtract
print(subtract(5, 2))

from math_utils import *
print(add(1, 1))  # Only works if __all__ is defined
```

### `__all__` in `__init__.py`

```python
__all__ = ["add", "subtract", "area_circle"]
```

This controls what gets imported with `from math_utils import *`.

# ðŸ”¹ `__all__` in `__init__.py`

The special variable **`__all__`** is a list of strings that defines the **public API** of a module or package.
It explicitly controls what is imported when someone uses:

```python
from package_name import *
```

---

## ðŸ§© Example: `math_utils` package

Suppose we have this package structure:

```
math_utils/
â”‚
â”œâ”€â”€ __init__.py
â”œâ”€â”€ basic_ops.py
â””â”€â”€ geometry.py
```

### **basic\_ops.py**

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def _secret_formula(x):   # intended to be private
    return x * 42
```

### **geometry.py**

```python
import math

def area_circle(r):
    return math.pi * r * r

def area_square(a):
    return a * a
```

### **`__init__.py`**

```python
from .basic_ops import add, subtract
from .geometry import area_circle

__all__ = ["add", "subtract", "area_circle"]
```

---

## ðŸš€ How it works

Now, when you import like this:

```python
from math_utils import *
```

âœ… Only these names are available:

```python
add
subtract
area_circle
```

âŒ Hidden:

* `area_square` is not imported.
* `_secret_formula` is not imported.

---

## âš–ï¸ Without `__all__`

If you **donâ€™t define `__all__`**, Python will import **all names that do not start with an underscore (`_`)**.

So in our example:

```python
from math_utils import *
```

would import `add`, `subtract`, `area_circle`, **and** `area_square`.

---

## ðŸŽ¯ Why use `__all__`?

1. **Encapsulation** â†’ hide helper functions (`_secret_formula`).
2. **Clean API** â†’ only expose the important functions.
3. **Documentation** â†’ makes it clear whatâ€™s public.
4. **Avoid Conflicts** â†’ prevents polluting the namespace with too many imports.

---

## ðŸ“ Tips & Tricks

* Use `__all__` at the **package level** (`__init__.py`) to control what the package exports.
* Use it at the **module level** (`.py` files) if you want fine-grained control inside a module.
* Prefix helper functions with `_` and leave them out of `__all__`.
* Tools like **Sphinx** (for docs) respect `__all__` â†’ it defines what shows up as "public".

---

âœ… In short:
`__all__` is your way of saying *"these are the official functions and classes I want people to use from my package."*

## **5. Nested (Sub) Packages**

You can have sub-packages inside packages:

```
data_science/
    __init__.py
    statistics/
        __init__.py
        mean.py
```

**statistics/mean.py**

```python
def mean(values):
    return sum(values) / len(values)
```

**Usage:**

```python
from data_science.statistics.mean import mean
print(mean([1, 2, 3, 4]))  # 2.5
```

# ðŸ”¹ What does **pip** mean?

* **pip** is a **recursive acronym** (the letters refer back to itself).
* It stands for: **"Pip Installs Packages"** (or sometimes jokingly **"Pip Installs Python"**).

That means its **only job** is to install and manage **Python packages** (libraries, modules, or frameworks).

---

# ðŸ”¹ What are Python packages?

A **package** in Python is basically a **collection of Python modules** (code files) bundled together so developers can reuse them easily.

* Example: `requests` â†’ package for making HTTP requests.
* Example: `flask` â†’ package for building web apps.
* Example: `numpy` â†’ package for numerical computing.

Instead of everyone writing their own HTTP client, you just **install** `requests` with pip.

---

# ðŸ”¹ Where does pip get packages from?

By default, pip installs from **PyPI (Python Package Index)** â†’ the official repository of Python software.

* URL: [https://pypi.org](https://pypi.org)
* It has **over 500,000 packages** (as of 2025).

You can also configure pip to install from:

* A **private repository** (like an internal company package index).
* A **local file** (wheel or tarball).
* A **GitHub repo**.

---

# ðŸ”¹ Why is pip important?

Before pip existed, installing packages in Python was painful:

* You had to download source code manually.
* Compile or copy it into your project.
* Manage versions by yourself.

With pip:
âœ… One command installs everything
âœ… Handles dependencies (installs other required packages automatically)
âœ… Works with virtual environments for isolation

---

# ðŸ”¹ Quick Example

Letâ€™s say you want to use Flask for a web app:

```bash
pip install flask
```

What happens?

1. Pip contacts **PyPI**.
2. Finds the latest version of Flask.
3. Downloads the package files.
4. Installs Flask **and all dependencies** (like `Werkzeug`, `Jinja2`, etc.).
5. Makes it available in your Python environment.

Now you can use it:

```python
from flask import Flask

app = Flask(__name__)
```

# ðŸ”¹ Summary

* **pip** = tool for installing/managing Python packages.
* Default source = **PyPI**.
* Simplifies dependency management.
* Essential for almost all Python development.
